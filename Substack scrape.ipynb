```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get('https://defieducation.substack.com')
```

```python
def sub_driver():
    driver = webdriver.Chrome(service=service)
    try:
        driver.get('https://defieducation.substack.com')
    except Exception as e:
        driver.quit()
        print('in sub_driver() error: ', e)
        time.sleep(3)
        return sub_driver()
        

    # go to login page
    login = driver.find_elements(By.CSS_SELECTOR, 'a[href="https://substack.com/sign-in?redirect=%2F&for_pub=defieducation"]')
    login[1].click()

    #log in with email and password
    driver.find_element(By.CSS_SELECTOR, 'a[href="javascript:void(0)"').click()
    email = driver.find_element(By.CSS_SELECTOR, 'input[placeholder="Your Email"]')
    password = driver.find_element(By.CSS_SELECTOR, 'input[placeholder="Your Password"]')
    email.send_keys(os.environ["EMAIL"])
    password.send_keys(os.environ["SUBS"])
    driver.find_element(By.XPATH,"//button[text()='Sign in']").click()
    time.sleep(1)
    return driver
```
